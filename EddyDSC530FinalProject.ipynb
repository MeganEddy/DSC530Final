## Hypothesis

Does an increase in urban population lead to a faster yearly increase in temperatures?

Countries with higher urban populations have larger yearly increases in temperature.

## Variables

1. __Average Temperature__:
Average increase of temperature by year in degrees Celcius

2. __Urban Population__:
Number of people living in urban areas

3. __Rural Population__:
Number of people living in suburban areas

4. __Total Emissions__:
Total greenhouse gas emissions (CO2) from various sources
* Measured in kilotons (kt) - 1 kt = 1000kg of CO2

5. __Food Retail__:
Emissions from retail establishments selling food
* Measured in kilotons (kt) - 1 kt = 1000kg of CO2

6. __Year__:
The data was collected from 1990 until 2020

## PMF

Compares the probability that the Average Temperature will actually remain the same/decrease vs. increase.

```python
from collections import Counter
```

```python
# Creates empty lists to store the values
lessthan = []
morethan = []

# Loops through the column Average_Temperature and adds each value to the
# appropriate list

for value in emissionsdf.Average_Temperature:
    if value <= 0:
        lessthan.append(value)
    else:
        morethan.append(value)
```

```python
# Finds the total number of data points
total_lessthan = len(lessthan)
total_morethan = len(morethan)

# Gets the frequency of each unique value
value_lessthan = Counter(lessthan)
value_morethan = Counter(morethan)

# Calculates the probability for each unique value
prob_lessthan = [count / total_lessthan for count in value_lessthan.values()]
prob_morethan = [count / total_morethan for count in value_morethan.values()]

# Get the corresponding unique values
unique_values_lessthan = list(value_lessthan.keys())
unique_values_morethan = list(value_morethan.keys())
```

```python
plt.bar(unique_values_lessthan, prob_lessthan, alpha=0.7, label="lessthan")
plt.bar(unique_values_morethan, prob_morethan, alpha=0.7, label="morethan", color='g')

plt.xlabel("Temperature Change")
plt.ylabel("Probability")
plt.title("Probability Mass Functions")
plt.legend()
plt.grid(False)
plt.show()
```

## Lognormal Distribution

```python
def MakeNormalModel(urbpop):
    """Plots a CDF with a Normal model.

    weights: sequence
    """
    cdf = thinkstats2.Cdf(urbpop, label="Urban Population")

    mean, var = thinkstats2.TrimmedMeanVar(urbpop)
    std = np.sqrt(var)
    print("n, mean, std", len(urbpop), mean, std)

    xmin = mean - 4 * std
    xmax = mean + 4 * std

    xs, ps = thinkstats2.RenderNormalCdf(mean, std, xmin, xmax)
    thinkplot.Plot(xs, ps, label="model", linewidth=4, color="0.8")
    thinkplot.Cdf(cdf)
```

```python
MakeNormalModel(log_urbpop)
thinkplot.Config(
    title="Urban Population, log scale",
    xlabel="Population(log10)",
    ylabel="CDF",
    loc="upper left",
)
```

```python
def MakeNormalPlot(urbpop):
    
    mean, var = thinkstats2.TrimmedMeanVar(urbpop, p=0.01)
    std = np.sqrt(var)

    xs = [-5, 5]
    xs, ys = thinkstats2.FitLine(xs, mean, std)
    thinkplot.Plot(xs, ys, color="0.8", label="Model")

    xs, ys = thinkstats2.NormalProbability(urbpop)
    thinkplot.Plot(xs, ys, label="Urban Population")
```

```python
MakeNormalPlot(log_urbpop)
thinkplot.Config(
    title="Urban Population, lognormal plot",
    xlabel="Urban Population (log10)",
    ylabel="CDF",
    loc="upper left",
)
```

## Scatterplots

### Plot 1: Urban Population versus Average Temperature

```python
plt.scatter(log_urbpop, tempdf, marker='o', color='skyblue', alpha = 0.25, s = 5)

plt.xlabel('Log Urban Population')
plt.ylabel('Average Temperature')
plt.title('Log Urban Pop vs. Average Temperature')

plt.show()
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
log_urbpop = np.log10(urbpop + 1e-10)
Covariance = Cov(urbpop, tempdf)
Log_covariance = Cov(log_urbpop, tempdf)

print("Covariance:", Covariance)
print("Log Covariance:", Log_covariance)

correlation = np.corrcoef(urbpop, tempdf)[0, 1]
print("Pearson's Correlation Coefficient:", correlation)
```

## Hypothesis Testing

### Correlation

```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
data = log_urbpop.values, tempdf.values
ht = CorrelationPermute(data)
pvalue = ht.PValue()
print(pvalue)

ht.actual, ht.MaxTestStat()
```
